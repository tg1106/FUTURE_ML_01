1. Imports

```python
import pandas as pd
import os
```

2. Load Raw Data

```python
df = pd.read_csv("../data/raw/train.csv")
df["date"] = pd.to_datetime(df["date"])

print("Raw shape:", df.shape)
df.head()
```

3. Aggregate Total Daily Sales

```python
daily_sales = df.groupby("date")["sales"].sum().reset_index()
daily_sales = daily_sales.sort_values("date").reset_index(drop=True)

print("Aggregated shape:", daily_sales.shape)
print("Date range:", daily_sales["date"].min(), "to", daily_sales["date"].max())
daily_sales.head()
```

4. Save Aggregated Data

```python
os.makedirs("../data/interim", exist_ok=True)

daily_sales.to_csv("../data/interim/daily_sales_aggregated.csv", index=False)
print("Saved daily_sales_aggregated.csv")
```

## 📌 Summary

- Loaded raw `train.csv` (~3M rows, per-store per-family)
- Aggregated into total daily sales (single time series)
- Sorted chronologically and saved to `data/interim/daily_sales_aggregated.csv`

### 🎯 Purpose
Reduce granular store-level data into a single daily time series suitable for forecasting.
