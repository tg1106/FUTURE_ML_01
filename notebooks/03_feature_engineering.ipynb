1. Imports

```python
import pandas as pd
import numpy as np
import os
```

2. Load Aggregated Data

```python
df = pd.read_csv("../data/interim/daily_sales_aggregated.csv")
df["date"] = pd.to_datetime(df["date"])

print("Shape:", df.shape)
df.head()
```

3. Time-Based Features

```python
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["day"] = df["date"].dt.day
df["dayofweek"] = df["date"].dt.dayofweek
df["weekofyear"] = df["date"].dt.isocalendar().week.astype(int)
df["is_weekend"] = df["dayofweek"].isin([5, 6]).astype(int)

df.head()
```

4. Lag Features

```python
df["lag_1"] = df["sales"].shift(1)
df["lag_7"] = df["sales"].shift(7)
df["lag_30"] = df["sales"].shift(30)

df.head()
```

5. Rolling Window Features

```python
df["rolling_7_mean"] = df["sales"].rolling(window=7).mean()
df["rolling_30_mean"] = df["sales"].rolling(window=30).mean()
df["rolling_7_std"] = df["sales"].rolling(window=7).std()

df.head()
```

6. Drop NA Rows

```python
print("Shape before dropna:", df.shape)
df = df.dropna().reset_index(drop=True)
print("Shape after dropna:", df.shape)
```

7. Save Feature Dataset

```python
os.makedirs("../data/processed", exist_ok=True)

df.to_csv("../data/processed/features.csv", index=False)
print("Saved features.csv")
print("Columns:", list(df.columns))
```

## 📌 Summary

- Loaded aggregated daily sales from `data/interim/daily_sales_aggregated.csv`
- Created time-based features: `year`, `month`, `day`, `dayofweek`, `weekofyear`, `is_weekend`
- Created lag features: `lag_1`, `lag_7`, `lag_30`
- Created rolling features: `rolling_7_mean`, `rolling_30_mean`, `rolling_7_std`
- Dropped NA rows caused by shifting/rolling
- Saved final ML-ready dataset to `data/processed/features.csv`

### 🎯 Purpose
Transform daily sales time series into structured numerical features suitable for machine learning models.
